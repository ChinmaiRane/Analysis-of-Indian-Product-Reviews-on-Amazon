**Analysis of Indian Product reviews**

1.   Acquire: Dataset was acquired from kaggle namely Indian Products on Amazon by Neha Prabhavalkar
2.   Parse:The data was in .csv form and was converted to dataframe for analysis and then again converted to .csv for using it in Tableau.
3.   Filter: Filters were applied for considering top five brands with most and least reviews,bar chart and wordcloud in tableau by just dragging the reviews column to filter section and then setting values inside it.
4.   Mining: Sentiment analysis was done to identify emotions in the text so that it can be used in visualizing key reviews of products.
5.   Represent: Data is being represented using various types of graphs such as bar chart,treemaps,word cloud,pie chart,packed bubbles which were decided according to the data being mapped.
6.   Refining:Data refinement was done in Tableau before visualization which included checking uppercase and lowercase in names of brand and making the brand names shorter ad easier to read by removing irrelevant parts from the name.
7.   Interactive:All the data visualizations done in Tableau are very interactive and give breif information about the brand when we hover over it.
